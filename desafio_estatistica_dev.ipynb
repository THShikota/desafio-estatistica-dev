```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
        ],
    'valor': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
        ]
}
```

```python
dict_faturamento
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
df_faturamento.info()
```

```python
df_faturamento['mes'] = df_faturamento['data_ref'].dt.month
df_faturamento
```

```python
df_faturamento['valor'].describe()
```

```python
# Trazer as médias das vendas
df_faturamento['valor'].mean()
```

```python
# Criar um gráfico de barras vertical mostrando o mês de referência e o valor
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt

ax = df_faturamento.plot.bar(x='mes', y='valor', figsize=(10, 6))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%d'))
plt.title('Faturamento por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor de Faturamento')
plt.xticks(rotation=0)

# Adicionar os valores acima de cada barra
for p in ax.patches:
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(
        f"{int(y)}",
        (x, y),
        ha="center",
        va="bottom",
        xytext=(0, 3),
        textcoords="offset points",
        clip_on=False
    )

plt.tight_layout()
plt.show()
```

```python
# Criar um gráfico de linhas mostrando o mês de referência e o valor.

ax = df_faturamento.plot.line(x='mes', y='valor', figsize=(8, 5))
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%d'))
plt.title('Faturamento por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor de Faturamento')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```
